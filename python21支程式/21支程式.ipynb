```python
#1 #記帳本
income1 = 0 #薪水總額
income2 = 0 #兼職總額
income3 = 0 #其他收入總額
incomeall = 0 #總收入
outlay1 = 0 #飲食總額
outlay2 = 0 #交通總額
outlay3 = 0 #娛樂總額
outlay4 = 0 #其他支出總額
outlayall = 0 #總支出
balance = 0 #餘額
print('歡迎使用記帳本')
while 1:
    select = input("請選擇要記帳的內容1.收入2.支出0.結束:")
    if select == '1':
        typein = input('請選擇類型1.薪水2.兼職3.其他:')
        if typein == '1':
            income = eval(input("請輸入金額:"))
            income1+=income
        if typein == '2':
            income = eval(input("請輸入金額:"))
            income2+=income    
        if typein == '3':
            income = eval(input("請輸入金額:"))
            income3+=income 
    if select == '2':
        typein = input('請選擇類型1.飲食2.交通3.娛樂4.其他:')
        if typein == '1':
            outlay = eval(input("請輸入金額:"))
            outlay1+=outlay
        if typein == '2':
            outlay = eval(input("請輸入金額:"))
            outlay2+=outlay    
        if typein == '3':
            outlay = eval(input("請輸入金額:"))
            outlay3+=outlay
        if typein == '4':
            outlay = eval(input("請輸入金額:"))
            outlay4+=outlay
    if select == '0':
        break
incomeall=income1+income2+income3 #計算總收入
outlayall=outlay1+outlay2+outlay3+outlay4 #計算總支出
print('收入明細:\n薪水:{:d}:兼職{:d}:其他:{:d}'.format(income1,income2,income3))
print('支出明細:\n飲食:{:d}:交通{:d}:娛樂:{:d}其他:{:d}'.format(outlay1,outlay2,outlay3,outlay4))
print('總收入:{:d}總支出:{:d}'.format(incomeall,outlayall))
if incomeall>outlayall:
    balance=incomeall-outlayall
    print('存了{:d}元呢真棒!'.format(balance))
if incomeall<outlayall:
    balance=outlayall-incomeall
    print('負債{:d}元喔要多賺錢了!'.format(balance))
if incomeall==outlayall:
    print('剛好打平呢')
```

```python
#12 #俄羅斯輪盤
import random
money=1000
bonus=0
while 1:
    play=input("要玩俄羅斯輪盤嗎:1.好2.不要:")#取一個值判斷是否要玩
    if play=="2":
        print('你可以拿走{:d}元'.format(money))
        break
    x=random.randint(0,36) #隨機取得一數
    if x==0:          #設定判斷用變數
        colour='綠色'
        z='偶數'
    if x%2==0:
        colour='紅色'
        z='偶數'
    if x%2==1:
        colour='黑色'
        z='奇數'
    if 1<=x<=12:
        path='前'
    if 13<=x<=24:
        path='中'
    if 25<=x<=36:
        path='後'
    odds=1 #賠率
    pay=eval(input('請輸入要下注的金額:'))
    select=input('請選擇下注的方式:1.數字2.顏色3.大小4.奇偶數5.前中後:')
    if select=='1': #數字的迴圈
        player=eval(input('請問輸入你選擇的數字(最多6個):'))
        number=0 #判斷for迴圈跑了幾次藉此確認所有號碼皆未中獎
        type1=str(type(player)) #將輸入的狀態轉為str存入type1
        length=len(type1)  #藉由type1的長度判斷輸入的是list還是int
        if length==13: #int時也是輸入1個數字時跑的
            odds=35
            if x==player:
                bonus=pay*odds #將賭注*賠率所獲的的獎金
                money+=bonus   #將剩下的錢加上得到的錢
                print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money)) 
            if x!=player:
                money-=pay
                print('還剩{:d}元'.format(money))
        if length==15:  #list時也是輸入多個數字時跑的
            if len(player)==2: #設定賠率
                odds=17
            if len(player)==3:
                odds=11
            if len(player)==4:
                odds=8
            if len(player)==5:
                odds=5
            if len(player)==6:
                odds=3
            for i in player:
                number+=1
                if i==x:
                    bonus=pay*odds
                    money+=bonus
                    print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
                    break
                if number==len(player) and i!=x:
                    money-=pay
                    print('還剩{:d}元'.format(money))
    if select=='2':#顏色的迴圈
        player=input('請問選擇你要的顏色(紅色、黑色、綠色):')
        if colour=='紅色' and player=='紅色':
            bonus=pay*odds
            money+=bonus
            print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
        if colour=='黑色' and player=='黑色':
            bonus=pay*odds
            money+=bonus
            print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
        if colour=='綠色' and player=='綠色':
            odds=50
            bonus=pay*odds
            money+=bonus
            print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
        if colour!='紅色' and player=='紅色':
            money-=pay
            print('還剩{:d}元'.format(money))
        if colour!='黑色' and player=='黑色':
            money-=pay
            print('還剩{:d}元'.format(money))
        if colour!='綠色' and player=='綠色':
            money-=pay
            print('還剩{:d}元'.format(money))
    if select=='3':#大小的迴圈
        player=input('請問選擇大小:')
        if z=='大' and player=='大':
            bonus=pay*odds
            money+=bonus
            print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
        if z=='小' and player=='小':
            bonus=pay*odds
            money+=bonus
            print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
        if z!='大' and player=='大':
            money-=pay
            print('還剩{:d}元'.format(money))
        if z!='小' and player=='小':
            money-=pay
            print('還剩{:d}元'.format(money))
    if select=='4':#奇偶數的迴圈
        player=input('請問選擇奇偶數:')
        if z=='奇數' and player=='奇數':
            bonus=pay*odds
            money+=bonus
            print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
        if z=='偶數' and player=='偶數':
            bonus=pay*odds
            money+=bonus
            print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
        if z!='奇數' and player=='奇數':
            money-=pay
            print('還剩{:d}元'.format(money))
        if z!='偶數' and player=='偶數':
            money-=pay
            print('還剩{:d}元'.format(money))
    if select=='5':#前中後的迴圈
        player=input('請選擇前中後:')
        odds=2
        if path=='前' and player=='前':
            bonus=pay*odds
            money+=bonus
            print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
        if path=='中' and player=='中':
            bonus=pay*odds
            money+=bonus
            print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
        if path=='後' and player=='後':
            bonus=pay*odds
            money+=bonus
            print('恭喜你中獎共獲得:{:d}元 你還剩{:d}元'.format(bonus,money))
        if player=='前' and path!='前':
            money-=pay
            print('還剩{:d}元'.format(money))
        if player=='中' and path!='中':
            money-=pay
            print('還剩{:d}元'.format(money))
        if player=='後' and path!='後':
            money-=pay
            print('還剩{:d}元'.format(money))
    if money==0:
        print('你沒錢了')
        break
    if money<0:
        print('你欠債:{:d}元'.format(money))
        break
```

```python
#14 #隨機數排序
import random #取得姓名
import random as r
a1=['林','陳','李','王','張']
a2=['怡','欣','雅','心','志','家','志','俊','佩']
a3=['君','怡','雯','豪','婷','玲','惠','偉','宏']
x=list()
y=list()
for i in range(0,10):
    x.append(random.randint(45,90)) #取得體重
    y.append(random.randint(155,180)) #取得身高
x.sort()
y.sort()
print(x)
print(y)
for i in range(0,10):
    name=r.choice(a1)+r.choice(a2)+r.choice(a3) #在序列中隨機抓一個
    BMI=x[i]/(y[i]/100)**2
    if BMI<18.5:
        z='太輕'
    if 18.5<=BMI<24:
        z='標準'
    if 24<=BMI<27:
        z='過重'
    print(name+'身高:{:d} 體重:{:d} BMI:{:.1f} {:s}'.format(y[i],x[i],BMI,z))
```

```python
#16 #投資計算機
def money(in1,year):
    if year<=0.5:
        rate=0.35
    elif 0.5<year<=1:
        rate=0.65
    elif 1<year<=2:
        rate=0.755
    elif 2<year<=3:
        rate=0.76
    elif 3<year:
        rate=0.785
    money=in1+in1*year*rate
    print("你的本金為:{:d}".format(in1))
    print("你存放:{:d}年".format(year))
    print("你的利率為:{:.3f}".format(rate))
    print("你的利息為:{:.0f}".format(in1*year*rate))
    print("本金加利息你共獲得{:.0f}元".format(money))
in1=eval(input("請輸入你要存放多少錢:"))
year=eval(input("請輸入你要存放多少年:"))
money(in1,year)
```

```python
#17 #遞減平方相加器
def sum1(number):
    sum1=0
    for i in range(number,0,-1):
        x=i*i
        sum1=x+sum1
        print("{:d}".format(x),end='')
        if i==1:
            print("=",end='')
        else:
            print("+",end='')
    print(sum1)
number=eval(input("請輸入一個數字:"))
sum1(number)
```

```python
#18 #字典查詢機
dict1={}
z=0
def new(x,y):
    dict1[x]=y
    print("成功新增單字:"+x+"中文解釋:"+y)
def dell(d):
    if(f in dict1):
        del dict1[d]
        print("成功刪除{:s}".format(f))
    else:
        print("查無此字")
def find(f):
    if(f in dict1):
        print('你查詢的單字是:'+f+"中文解釋是:"+dict1[f])
    else:
        print("查無此字")
def clear():
    dict1.clear()
    print("已清除全部單字")
print("歡迎使用字典")
while z!='5':
    z=input("請問你要使用甚麼功能:\n1.新增\n2.刪除\n3.查詢\n4.刪除整本\n5.離開\n")
    if z=='1':
        x=input("請輸入英文單字:")
        y=input("請輸入中文解釋:")
        new(x,y)
    if z=='2':
        d=input("請輸入要刪除的單字:")
        dell(d)
    if z=='3':
        f=input("請輸入要查找的單字:")
        find(f)
    if z=='4':
        clear()
```
